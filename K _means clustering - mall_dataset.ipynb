### K-Means clustering

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
#extracting only 3rd and 4th features.bcz we need a 2D plot
x=data.iloc[:,[3,4]].values
```

### Finding the optimal number of clusters using the elbow method

#### The elbow method uses WCSS concept to de=raw the plot by plotting WCSS values on Y axis 

```python
# finding optimal number of clusters using the elbow method
from sklearn.cluster import KMeans

wcss_list=[] #initialize the list for tghe value of WCSS

#using for loop for iterations from 1 to 10
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42,n_init=10)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)  #inertia model
plt.plot(range(1,11),wcss_list)
plt.title('The elbow method graph')
plt.xlabel('number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()          # elbow -  optimum numbr of clusters
```

### Training the K-means algorithm on the training dataset

```python
# training the k-means model on a dataset
kmeans=KMeans(n_clusters=2,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
```

```python
# training the k-means model on a dataset
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
```

### Visualizing the clusters

```python
plt.scatter(x[y_predict == 0,0],x[y_predict == 0,1],s=100,c='blue',label='cluster1') #for 1st cluster
plt.scatter(x[y_predict == 1,0],x[y_predict == 1,1],s=100,c='green',label='cluster2') #for 2nd cluster
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroid')
plt.title('clusters of customers')
plt.xlabel('annual income(k$)')
plt.ylabel('spending score (1-100)')
plt.legend()
plt.show()
```

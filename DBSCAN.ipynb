### DBSCAN - clustering

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
x=data.iloc[:,[3,4]].values
```

### Minpoints

### Epsilon

### compute data proximity from each other using nearest neighbors

```python
from sklearn.neighbors import NearestNeighbors #library import
neighb=NearestNeighbors(n_neighbors=2)   #crtng an obj of the nearestneighbors clss
nbrs=neighb.fit(x)                    #fitting the data to the obj
distances,indices=nbrs.kneighbors(x)   #finding the nearest neighbors
```

### Sorting and plot the distances between the data points

```python
#sort and plot the distances results
distances=np.sort(distances,axis=0) #sorting the distances
distances=distances[:,1] #taking the 2nd column of the sorted distances
plt.rcParams['figure.figsize']=(10,8) #setting figure size
plt.plot(distances)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
#cluster the data into five clusters
dbscan=DBSCAN(eps=8,min_samples=4).fit(x) #model fitting
labels=dbscan.labels_ #get labels
```

```python
#plot clusters
plt.scatter(x[:,0],x[:,1],c=labels,cmap='plasma') 
plt.xlabel('Income')
plt.ylabel('Spending score')
plt.show()
```
